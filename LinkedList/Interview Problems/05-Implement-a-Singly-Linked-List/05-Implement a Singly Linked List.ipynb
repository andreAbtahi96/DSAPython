# Implement a Singly Linked List

For this interview problem, create a node class and show how it can be used to create a Singly Linked List

```python
class Node:
    
    # constructor
    def __init__(self, value):
        self.value = value
        self.next = None

    
    def getValue(self):
        return self.value
```
